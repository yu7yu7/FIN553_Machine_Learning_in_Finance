```python
ones = np.ones_like(x1)
X = np.column_stack([ones, x1, x2])
Θ = np.zeros([3])

def p(Θ):
  return jax.nn.sigmoid(X @ Θ)


def L(Θ):
  return (- y * np.log(p(Θ)) - (1 - y) * np.log(1 - p(Θ))).mean()

grad_L = jax.grad(L)

def update(Θ):
  grad = grad_L(Θ)
  α = 0.1
  Θ = Θ - α * grad
  return Θ


for iteration in range(1000):
    Θ = update(Θ)
    print(Θ)
```

```python
# Plot predictions
plt.plot(y, marker='x', lw=0, markersize=10)
prediction = p(Θ) > 0.5
plt.plot(prediction, marker='o', lw=0, markersize=5)
```
