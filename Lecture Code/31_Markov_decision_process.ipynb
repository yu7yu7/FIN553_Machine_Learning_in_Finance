# Simulating the Markov Decision Process

A markov decision process(MDP) is a Markov reward process with decision. 

In different states and after different actions, we will have different probability and reward. (其實可以想像成有兩個不同的state系統) 

```python
import numpy as np

# Discount factor （折現率）
γ = 0.95

# Transition matrix if dont' invest
P_xinvest = np.array([[1, 0, 0],
              [0.1, .75, 0.15],
              [0.05, .1, 0.85]])

R_xinvest = np.array([0, 1, 2])

# Transition matrix if invest
P_invest = np.array([[1, 0, 0],
                     [0.05, .75, 0.2],
                     [0.02, .06, 0.92]])
R_invest = np.array([0, 0, 1])
```

## List Correspoding Policy

### Policy 1 

```python
# %% Policy # 0
#The policy never invest

π = [0, 0, 0]
# Implied Reward process
Rπ = np.array([0, 1, 2]) #看一看一看ㄧ

# Implied transition matrix
Pπ = np.array([[1, 0, 0],
               [0.1, .75, 0.15],
               [0.05, .1, 0.85]])

# Solution of the Bellman equation
I = np.eye(3)
vπ = np.linalg.solve(I - γ * Pπ, Rπ)
print(vπ)

```

###  Policy 2

```python
# %% Policy # 1
#Only invest in the good state
π = [0, 0, 1]

# Implied Reward process
Rπ = np.array([0, 1, 1]) #看一看一看二

# Implied transition matrix
Pπ = np.array([[1, 0, 0],
               [0.1, .75, 0.15],
               [0.02, .06, 0.92]])

# Solution of the Bellman equation
I = np.eye(3)
vπ = np.linalg.solve(I - γ * Pπ, Rπ)
print(vπ)

```
