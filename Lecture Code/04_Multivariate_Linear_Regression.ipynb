```python
import numpy as np
import matplotlib.pyplot as plt
K = np.array([35.48813504, 37.15189366, 36.02763376, 35.44883183, 34.23654799,
              36.45894113, 34.37587211, 38.91773001, 39.63662761, 33.83441519,
              37.91725038, 35.2889492, 35.68044561, 39.25596638, 30.71036058,
              30.871293, 30.20218397, 38.32619846, 37.78156751, 38.70012148,
              39.78618342, 37.99158564, 34.61479362, 37.80529176, 31.18274426])

T = np.array([1.06996051, 0.82167664, 1.22233446, 1.01092416, 0.95733097,
              0.88227781, 1.13711684, 0.97807517, 1.03421697, 0.7593949,
              1.05881775, 1.05604786, 1.058467, 1.22187404, 1.09091015,
              0.92975395, 0.96851598, 1.0988156, 0.78011274, 1.08338336,
              1.08531893, 0.85519128, 0.81446315, 0.90771418, 0.93185539])


y = np.array([4.38520061, 3.16149389, 4.67530465, 4.21746519, 4.40839628,
              3.53589972, 4.91069181, 3.24098257, 3.24996869, 3.8723874,
              3.71471908, 4.40032668, 4.29654058, 3.86564066, 6.03192935,
              5.48225334, 5.86622203, 3.73386091, 2.8726241, 3.60174454,
              3.36552398, 3.06984857, 3.8249487, 3.28088958, 5.37002006])

X = np.column_stack([np.ones_like(K), K, T])


def f(X, Θ):
    return X @ Θ


def L(Θ):
    e = f(X, Θ) - y
    N = len(y)
    return 0.5 * e.T @ e / N


def grad_L(Θ):
    N = len(y)
    return X.T @ (X @ Θ - y) / N


def apply_gradients(Θ, grads):
    α = 1e-3
    Θ = Θ -α * grads
    return Θ


Θ = np.array([-10., 0., 0.])

```

```python
for iteration in range(1500000):
  grads = grad_L(Θ)
  Θ = apply_gradients(Θ, grads)

print(Θ)

fitted_value = f(X, Θ)
plt.figure(figsize=[12, 8])
plt.scatter(K, y, s=140)
plt.xlabel('Strike price')
plt.ylabel('Price of call option')
plt.scatter(K, fitted_value, color='k')
plt.legend(['Observed values', 'Fitted values'])
```
