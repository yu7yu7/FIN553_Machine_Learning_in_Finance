```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)

data = np.load('data.npy')
Θ = data[:, 0]
time = data[:, 1]
T = len(time)

f = plt.figure()
f.set_figwidth(10)
f.set_figheight(4)
plt.plot(time, Θ, lw=0.5)
plt.xlabel('time (days)')
plt.ylabel('stock price Θ')

```

```python
V = np.zeros_like(Θ)
β = 0.9
for t in range(T - 1):
    V[t + 1] = β * V[t] + (1 - β) * Θ[t]
    print(V)

f = plt.figure()
f.set_figwidth(15)
f.set_figheight(4) 
plt.plot(time, Θ, lw=0.5, )
plt.plot(time, V)
```
