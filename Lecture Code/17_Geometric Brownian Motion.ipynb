# Create the Brownian Motion

```python
import numpy as np
import matplotlib.pyplot as plt

μ = 0.06  #
σ = 0.1
dt = 0.1


# var(dZ) = dt
# var(ε) = 1.

# var(np.sqrt(dt) * ε) =
# (np.sqrt(dt))**2 * var(ε)
# dt * 1 = dt


def my_function(S):
    dZ = np.random.normal(scale=np.sqrt(dt))
    dS = μ * S * dt + σ * S * dZ
    return dS + S


T = 100 #time of simulation
S_path = np.zeros(T)
S_path[0] = 100 #seed
for t in range(T - 1):
    S_path[t + 1] = my_function(S_path[t])

plt.plot(S_path)
plt.xlabel('Period')
plt.ylabel('Stock Price')
plt.title('Geometric Brownian Motion')
```
